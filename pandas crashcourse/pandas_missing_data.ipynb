```python
import numpy as np
import pandas as pd
```

### 3 things you can do with missing data:
### - Leave it as missing
#### * use this mostly with categorical data
### - Remove the missing data
#### * if you're missing a large percentage of data it's best to remove the column
#### * if you're missing a small percentage of data it's best to remove that row
### - Fill in the missing data
#### * if the missing data is important then replace it with a reasonable value

```python
df = pd.DataFrame({
    'A': [1, 2, np.nan, 4],
    'B': [5, np.nan, np.nan, 8],
    'C': [10, 20, 30, 40]
})
```

```python
df
```

#### Remove the missing data:
.dropna()
##### you will get the rows where nothing is missing because the default value for axis is 0

```python
df.dropna()
```

#### if you set the axis to 1 you will get the cols where nothing is missing

```python
df.dropna(axis=1)
```

#### thresh determines how many NaN
##### - on the example below the threshhold is 3, so must have at least 2 NaN values


```python
df.dropna(axis=1, thresh=3)
```

##### - usually you calculate thresh based on a percentage
###### in example below we get back the cols with at least 75% of the values

```python
thresh = 0.25 * len(df)
df.dropna(axis=1, thresh=thresh)
```

#### Fill in the missing data

##### if the fill value is a different data type then the rest of the col values are converted to that data type.
###### on example below 1.0, 2.0, 4.0 5.0, 8.0 are now type string

```python
df.fillna(value='FILL VALUE')
```

```python
df.fillna(value=0)
```

#### You don't have to replace values in the entire df.

```python
df['A'] = df['A'].fillna(value=0)
```

```python
df
```

#### Fill in the values in 'B' with the avridge values of non missing elements

```python
df['B'].fillna(df['B'].mean())
```

```python
df.fillna(df.mean())
```
