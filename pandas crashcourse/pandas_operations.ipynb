```python
import numpy as np
import pandas as pd
```

```python
df_one = pd.DataFrame({'k1':['A','A','B','B','C','C'],
                      'col1':[100,200,300,300,400,500],
                      'col2':['NY','CA','WA','WA','AK','NV']})
```

#### Information on unique values

```python
df_one['col2'].unique()
```

##### .nunique() returns number of unique values

```python
df_one['k1'].nunique()
```

##### .value_counts() returns counts per unique entry

```python
df_one['col2'].value_counts()
```

##### .drop_duplicates() removes all duplicate values

```python
df_one.drop_duplicates()
```

#### Using functions to apply to all items
##### !Watch out for datatype errors!

```python
def grab_first_letter(state):
    return state[0]
```

```python
df_one['First letter'] = df_one['col2'].apply(grab_first_letter)
```

```python
df_one
```

```python
def complex_letter(state):
    if state[0] == 'W':
        return "Washington"
    else:
        return "Error"
```

```python
df_one['col2'].apply(complex_letter)
```

### Dummy variable features

```python
features = pd.DataFrame({
    'A': [100, 200, 300, 400, 500],
    'B': [12, 13, 14, 15, 16],
})
predictions = pd.DataFrame({'pred': [0, 1, 1, 0, 1]})
```

```python
features
```

```python
predictions
```

#### Concat two dataframes.
##### By default it concatinates along the rows. To concat along the columns set axis to 1

```python
pd.concat([features, predictions])
```

```python
pd.concat([features, predictions], axis=1)
```
