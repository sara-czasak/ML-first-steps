```python
import numpy as np
import pandas as pd
```

```python
columns = ['W', 'X', 'Y', 'Z']
```

```python
index = ['A', 'B', 'C', 'D', 'E']
```

```python
from numpy.random import randint
```

```python
np.random.seed(42)
data = randint(-100, 100, (5, 4))
```

```python
data
```

#### The pd.DataFrame() takes the data the index and the columns as params

```python
df = pd.DataFrame(data, index, columns)
```

```python
df
```

#### Getting a single column

```python
df["W"]
```

#### Getting multiple columns by passing in a list of column names. The column names don't have to be in order

```python
df[['Z', 'W']]
```

#### Create new column based on other columns (Feature Engineering)

```python
df['new'] = df['W'] + df['Y']
```

```python
df
```

#### Removing a column. Axis 1 means look for the column. Axis 0 means look for the row. df.drop doesn't remove anything permanently. To keep the change you need to reassign.

```python
df = df.drop('new', axis=1)
```

```python
df
```

#### To locate a row use the method .loc, it returns the row as a series. To locate multiple rows pass in a list.

```python
df.loc[['A', 'B']]
```

#### To locate by int index location use .iloc

```python
df.iloc[0:3]
```

#### To drop a row pass in the name of the row, you don't need to specify the axis since the default axis is 0

```python
df.drop('C')
```

#### Select a subset of rows and cols at the same time. You pass in the row and then the column

```python
df.loc[['A', 'C'], ['W', 'Y']]
```

```python
df
```

## Conditional selection

#### df comparison operator returns a df of bool values

```python
df > 0
```

#### If you pass in the condition to the df it returns df where the values that don't meet the condition are removed

```python
df[df > 0]
```

```python
df
```

```python
df['X'] > 0
```

#### Returns only rows where 'X' is greater than 0

```python
df[df['X'] > 0]
```

#### You can select on multiple conditions. Operators:
##### & and
##### | or

```python
df[(df['W']>0) & (df['Y']>1)]
```

#### Reset index to int values
##### .reset_index() creates int index col and converts old index col into a new col named index

```python
df.reset_index()
```

#### Setting existing column as index

```python
new_ind = ['CA', 'NY', 'WY', 'OR', 'CO']
```

```python
df['States'] = new_ind
```

#### .set_index gets rid of the old index and replaces it with the column that is passed in.
##### !REMEMBER! the index column name is not the name of the column.
###### In the example below the columns are 'W', 'X', 'Y', 'Z' and 'States' is the name of the index

```python
df = df.set_index('States')
```

```python
df
```

```python
df.columns
```
