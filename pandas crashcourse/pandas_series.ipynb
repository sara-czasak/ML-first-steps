```python
import numpy as np
import pandas as pd
```

```python
labels = ['a', 'b', 'c']
mylist = [10, 20, 30]
arr = np.array([10, 20, 30])
d = {'a': 10, 'b': 20, 'c': 30}
```

#### Create pandas data series from list setting another list as labels

```python
pd.Series(data=mylist, index=labels)
```

#### Pandas series from array

```python
pd.Series(arr, labels)
```

#### Pandas series from dictionary

```python
pd.Series(d)
```
